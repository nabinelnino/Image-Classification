```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
from tensorflow.keras.applications import imagenet_utils
from sklearn.metrics import confusion_matrix
import itertools
import os
import shutil
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

We first make a call to tf.keras.applications.mobilenet.MobileNet() to obtain a copy of a single pretrained MobileNet with weights that were saved from being trained on ImageNet images. We're assigning this model to the variable mobile.

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
def prepare_image(file):
    img_path = 'C:/Users/elnin/Desktop/WebAPI/Image-Classification/'
    img = image.load_img(img_path + file, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return tf.keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

```python
from IPython.display import Image
Image(filename='3.jpg', width=300,height=200) 
```

```python
preprocessed_image = prepare_image('3.jpg')
predictions = mobile.predict(preprocessed_image)
```

```python
results = imagenet_utils.decode_predictions(predictions)
```
